```python
from git import Repo, GitCommandError
import os
import pandas as pd
import subprocess
import shutil
import sys
```

```python
def clone(gitUrl, repoDir, sample):
    try:
        Repo.clone_from(gitUrl, repoDir + sample, depth=1)
        subprocess.run(["git", "-C", repoDir + sample, "fetch", "--unshallow"], check=True)
        print("Repository has been successfully unshallowed.")
    except GitCommandError as e:
            print("An error occurred while downloading %s" % sample)
            print(f'Error executing git command {' '.join(["git", "-C", repoDir + sample, "fetch", "--unshallow"])}: {e}', file=sys.stderr)
```

```python
def removeRepo(sample):
    if os.path.isdir('repositories/'+sample):
        sampleDir = os.path.join('repositories/', sample)
        for root, dirs, files in os.walk(sampleDir, topdown=False):
                    for name in files:
                        os.chmod(os.path.join(root, name), 0o777)  # Change permissions to ensure file can be deleted
                    for name in dirs:
                        os.chmod(os.path.join(root, name), 0o777) # Change permissions to ensure directory can be deleted
        shutil.rmtree(sampleDir)
        print("Previous download deleted")
```

```python

def download(sample):
    gitHubUrl = f"https://github.com/{sample}.git"
    repoDir = "repositories/"
    if os.path.isdir(repoDir + sample):
        print("Repository %s already exists" % sample)
    else:
        print("Downloading %s" % sample)
        clone(gitHubUrl, repoDir, sample)
        print("%s downloaded" % sample)
        
```

```python
samples = pd.read_csv(os.path.join('..','1-GetQuestions','questions.csv'))
```

```python
len(samples.groupby(by="path", dropna=False))
```

```python
for i in range (0, len(samples)):
    print(f'percentage concluded: {i/len(samples):.2%}')
    sample = samples.iloc[i]
    sample_name = sample["framework"]+"/"+sample["path"]
    removeRepo(sample_name)
print(f'{"\033[92m"}All projects deleted {"\033[0m"}')
```

```python
sample_names = (samples["framework"] + "/" + samples["path"]).unique()
for i in range (0, len(sample_names)):
    print(f'percentage concluded: {i/len(sample_names):.2%}')
    download(sample_names[i])
print(f'{"\033[92m"}All projects downloaded{"\033[0m"}')
# for i in range (0, len(samples)):
#     print(f'percentage concluded: {i/len(samples):.2%}')
#     sample = samples.iloc[i]
#     sample_name = sample["framework"]+"/"+sample["path"]
#     download(sample_name)
# print(f'{"\033[92m"}All projects downloaded{"\033[0m"}')
```
